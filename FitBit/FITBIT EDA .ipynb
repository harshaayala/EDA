```python
#import relevant libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import os
```

```python
activity = pd.read_csv('FitBit data.csv') # importing the dataset
```

```python
activity.head()
```

```python
#checking if there are any missing values
import missingno as msno
msno.matrix(activity)
```

```python
activity.shape # checking the number of rows and columns in the dataset
```

```python
data=activity.copy()
```

```python
data.drop(["Id"],axis=1,inplace=True)
```

```python
data['year'] = pd.DatetimeIndex(data['ActivityDate']).year
data['month'] = pd.DatetimeIndex(data['ActivityDate']).month
data['date'] = pd.DatetimeIndex(data['ActivityDate']).day
```

```python
data.head(5)
```

```python
import matplotlib.pyplot as plt
# figure size
plt.figure(figsize=(15,8))

sns.boxplot(x='date',y='Calories',data=data)
```

```python
data['week']=pd.to_datetime(data.ActivityDate).dt.week
data['Year'] = pd.to_datetime(data.ActivityDate).dt.year
```

```python
data['ActivityDate'] = pd.to_datetime(data['ActivityDate']) # converting it to datetime
```

```python
data['day'] = data['ActivityDate'].dt.day_name() # converting the day of the week to the name of the day
```

```python
data.head(5)
```

```python
plt.figure(figsize=(15,8))
ax=sns.barplot(x='day',y='Calories',data=data,palette='rainbow')
ax.set_title('Barplot of calories by the day of the week')
```
